```python
import pandas as pd
import json
```

```python
filepath = "static/data/us.json"
```

```python
with open(filepath) as f:    
            json_data = json.load(f)
        
json_data
```

```python
csv_file_1 = "resources/energy_census_and_economic_data_US_2010-2014.csv"
energy_df = pd.read_csv(csv_file_1)
energy_df.head()
```

```python
gdp_df = energy_df[['State','GDP2014']]
gdp_df.head()
```

```python
gdp_df = gdp_df.set_index('State')
gdp_df.head()
```

```python
for i in json_data['features']:
    state = i['properties']['NAME']
    gdp = gdp_df.loc[state, :][0]
    i['properties']['GDP'] = gdp
    print(i['properties'])
    
        
```

```python
df = pd.DataFrame(json_data)
df.head()
```
